```python
import pandas as pd
import urllib.request
import time
import os.path as osp

url = "https://resource-cms.springernature.com/springer-cms/rest/v1/content/17858272/data/v4"
!curl https://resource-cms.springernature.com/springer-cms/rest/v1/content/17858272/data/v4 > ./data/list.xlsx
library_list = pd.read_excel("./data/list.xlsx")
library_list.head()
```

```python
def doi_to_pdfname(doi=library_list["DOI URL"][0]):
    pdf_name = doi.split("http://doi.org/")[1] + ".pdf"
    pdf_name = pdf_name.replace("/", "%2F") # urlencode
    return pdf_name

def make_filename(title, author, package, folder):
    title = title.replace(" ", "_")
    author = author.replace(" ", "_").replace(".","_")
    package = package.replace(" ", "_").replace(",", "_")
    path = "./" + folder + "/" + package 
    if not osp.exists(path):
        os.makedirs(path)
    return path + "/" + author + "-" + title + ".pdf"

DL_SERVER = "https://link.springer.com/content/pdf/"
doi_list = library_list["DOI URL"]
title_list  = library_list["Book Title"]
author_list = library_list["Author"]
package_name  = library_list["English Package Name"]

success, fails = [],[]
for i, item in enumerate(zip(doi_list, title_list, author_list, package_name)):
    doi, title, author, package = item
    
    filename = make_filename(title, author, package, "springer_covid_19_books")
    url = DL_SERVER + doi_to_pdfname(doi)
    print(i, filename, ". url:", url)
    
    try:
        urllib.request.urlretrieve(url, filename)
        success.append((url, filename))
    except:
        fails.append((url, filename))
    
    time.sleep(3)
```
